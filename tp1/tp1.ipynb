```python
# Exercice 3

# Q1

v1 = np.arange(1, 20, 2)

print(v1)

rng = np.random.default_rng()
v2 = rng.integers(low=25, high=70, size=10)

print(v2)

concatenated = np.concatenate((v1, v2))

print(concatenated)

split = np.reshape(concatenated, (4, 5))

print(split)

v1v2_rw = np.stack((v1, v2), axis=0)

print(v1v2_rw)

v1v2_cw = np.stack((v1, v2), axis=1)

print(v1v2_cw)

v1v2_cw = np.column_stack((np.arange(0, len(v1v2_cw), 1) , v1v2_cw))

print(v1v2_cw)
```

```python
# Q2

A = np.array([
	[1, 2], 
	[3, 4]])
B = np.array([
	[5, 10, 2], 
	[15, 7, 20]])

a_plat = np.concatenate((A.flatten(), B.flatten()))

print(a_plat)

np.random.shuffle(a_plat)

print(a_plat)
print(np.argmin(a_plat))
print(np.argmax(a_plat))

C = np.random.randint(10, size=(5, 3)) 

print(C)
print(np.argmin(C, axis=0))
print(np.argmax(C, axis=0))

print(np.transpose(a_plat))
print(np.transpose(C))
```
